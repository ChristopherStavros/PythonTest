```python
import pandas
import folium
from geopy.geocoders import ArcGIS

gis = ArcGIS()
loc = gis.geocode("italy") 
map = folium.Map(location=[loc.latitude, loc.longitude], zoom_start=6)

destinations = ["Venice, Italy", "Rome, Italy", "Sorrento, Italy", "Naples, Italy", "Capri, Italy"]
fg = folium.FeatureGroup(name="Destinations")
for d in destinations:
    loc = gis.geocode(d)
    fg.add_child(folium.CircleMarker(
    location = [loc.latitude, loc.longitude],popup=d, radius = 7, fill_color='green', fill=True,  color = 'green', fill_opacity=0.7))

map.add_child(fg)
map.add_child(folium.LayerControl())
display(map)
```
